Disparities are calculated as a ratio of a metric for a group of interest compared to a base group. In our case, you should set the follwoing groups as base rate: Caucasian / Male /  Age 25-45.

```python
b = Bias()
```

```python

base_groups = {'race': 'Caucasian', 'sex': 'Male', 'age_cat': '25 - 45'}
bdf = b.get_disparity_predefined_groups(xtab, original_df=df,
                                        ref_groups_dict=base_groups,
                                        alpha=0.05, check_significance=True,
                                        mask_significance=True)
bdf.style
```

```python
# View disparity metrics added to dataframe
bdf[['attribute_name', 'attribute_value'] +
     b.list_disparities(bdf) + b.list_significance(bdf)].style
```
